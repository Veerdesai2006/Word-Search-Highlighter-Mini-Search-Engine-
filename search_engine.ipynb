```python
import string
Search_word=input("Enter the word to search")
File_path=input("Input file name with .txt")
try:
    with open (File_path , 'r') as file:
        content= file.read()
        content=content.lower()

        Search_word=Search_word.lower().strip()
        word_list=content.split()
        count= 0
        Highlighted = ""
        MY_word=Search_word
        for word in word_list:
            clean_word = word.strip(string.punctuation)
            if clean_word.lower() == Search_word:
                Highlighted+=Search_word.upper()+" "
                    
                count = count +1 
            else :
                Highlighted+=word.lower() + " "
except Exception as ex:
    print("File not found \n")
    print(ex)
else:
    if count>0:
        
        print(f"Here {MY_word.upper()}  is coming {count} Times \n")
        print(f"Here I have highlighted '{MY_word.upper()}' making it capital Wherever  in sentence it is coming. \n")
        print(Highlighted)
    else:
        print(f"The Word : {MY_word.upper()}, is not available in file")
```
